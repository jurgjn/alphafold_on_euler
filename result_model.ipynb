```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.rcParams['font.size'] = '14'
```

```python
data = {}
for model in range(5):
    path = "../output/SRSF1/"
    pkl_file = "result_model_" + str(model + 1) + ".pkl"
    path_file = path + pkl_file
    
    with open(path_file, 'rb') as f:
        data[model + 1] = pickle.load(f)
```

```python
plt.figure(figsize=(10, 6), dpi=100)

for model in [1]:
    plt.plot(data[model]["plddt"], label=model)

plt.xlabel('residue nr.')
plt.ylabel('pLDDT score')
#plt.legend(loc=1,framealpha=0.9);
plt.legend(loc="upper left")
plt.savefig("pLDDT.png")
```
